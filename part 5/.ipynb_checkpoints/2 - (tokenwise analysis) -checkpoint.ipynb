```python
def normalize_deprel(deprel):
    # change for taking only before @
    deprel = deprel.replace("@", ":")
    deprels = deprel.split(":")
    deprel_main = deprels[0]
    if len(deprels) > 1:
        deprel_aux = deprels[1]
    else:
        deprel_aux = 'none'

    return deprel_main, deprel_aux
```

```python
import conllu
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_agregated_cvs = "/home/wran/Research/memoire/experimentations_notebooks/data/6_cvs/agregated_cvs.conllu"
```

```python
with open(path_agregated_cvs, "r", encoding="utf-8") as infile:
    result = conllu.parse(infile.read())
```

```python
rows_list= []
for sequence in result:
    n_head_fail = 0
    n_deprel_fail = 0
    for token in sequence:
        head_true = int(token['head'])
        head_pred = int(token['misc']['head_MST'])
        deprel_true , deprel_aux_true = normalize_deprel(token['deprel'])
        deprel_main_pred = token['misc']['deprel_main_pred']
        if head_true != head_pred:
            n_head_fail += 1
#         if (deprel_main_true, deprel_aux_true) == ('punct', 'none'):
#             print("break")
#             break
        if (deprel_main_true, deprel_aux_true) != (deprel_main_pred, deprel_aux_pred):
            n_deprel_fail+=1
            print((deprel_main_true, deprel_aux_true))
            print((deprel_main_pred, deprel_aux_pred))
            print()
#             print(sequence.serialize())
#             print(token['form'])
#             print(head_true)            
#             print(head_pred)
#             print()
        row_dict = {
            'sent_id': sequence.metadata['sent_id'],
            'form' : token['form'],
            'position': token['id'],
            'dist_synt': abs(token['id'] - token['head']),
            'sent_length': len(sequence),
            '_counter': 1,
            'deprel_true': "{}:{}".format(deprel_main_true, deprel_aux_true),
            'deprel_pred': "{}:{}".format(deprel_main_pred, deprel_aux_pred),
    #         'length_copy': len(sequence),
            'head_fail' : int((head_true != head_pred)),
            'deprel_fail' : int((deprel_main_true, deprel_aux_true) != (deprel_main_pred, deprel_aux_pred)),
                   }
        rows_list.append(row_dict)
df = pd.DataFrame(rows_list) 
```

```python
df_groupby = df.groupby(['deprel_true']).sum()
```

```python
df_groupby['deprel_ratio'] = df_groupby['deprel_fail']/df_groupby['_counter']
df_groupby['head_ratio'] = df_groupby['head_fail']/df_groupby['_counter']
df_groupby
```

```python
df_groupby = df.groupby(['length']).sum()
df_groupby['ratio_head'] = df_groupby['n_head_fail']/df_groupby['length_copy']
df_groupby['ratio_deprel'] = df_groupby['n_deprel_fail']/df_groupby['length_copy']
df_groupby
```

```python
x = df_groupby.reset_index()['length']
y1 = df_groupby['ratio_deprel']
y2 = df_groupby['ratio_head']
plt.figure(figsize=(15,8))
plt.plot(x, y1)
plt.plot(x, y2)
plt.title("wrong prediction probability for a token depending on the sentence length")
plt.legend(['deprel', 'head'])
plt.show()
```
